```python
day_seven_input = open('day_7_input.txt').read()
day_seven_list = day_seven_input.splitlines()
```

```python
day_seven_list
```

```python
pwd = '/'
all_directories = [pwd]
current_directory_nest = []
current_directory_string = pwd
directory_sizes = {'/': 0}
for line in day_seven_list:
    line_split = line.split()
    directory_sizes_key = '-'.join(current_directory_nest)
    if line_split[0] == 'dir':
        all_directories.append(line[4:])
    elif line_split[0] == '$':
        if line_split[1] == 'cd':
            if line_split[2] != '..':
                pwd = line_split[2]
                current_directory_nest.append(pwd)
                directory_sizes_key = '-'.join(current_directory_nest)
                if directory_sizes_key not in directory_sizes:
                    print(current_directory_nest)
                    directory_sizes[directory_sizes_key] = 0
            else:
                directory_sizes['-'.join(current_directory_nest[:-1])] += directory_sizes['-'.join(current_directory_nest)]
                current_directory_nest = current_directory_nest[:-1] 
                pwd = current_directory_nest[-1]
            

            # print(line)
            # print(all_directories)
    else:
        directory_sizes[directory_sizes_key] += int(line_split[0])

    print(line)
    print(pwd)
    print(current_directory_nest)
    print(directory_sizes_key)
    print(directory_sizes)
```
