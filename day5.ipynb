```python
crates_dict = {
    '1': ['H', 'T', 'Z', 'D'],
    '2': ['Q', 'R', 'W', 'T', 'G', 'C', 'S'],
    '3': ['P', 'B', 'F', 'Q', 'N', 'R', 'C', 'H'],
    '4': ['L', 'C', 'N', 'F', 'H', 'Z'],
    '5': ['G', 'L', 'F', 'Q', 'S'],
    '6': ['V', 'P', 'W', 'Z', 'B', 'R', 'C', 'S'],
    '7': ['Z', 'F', 'J'],
    '8': ['D', 'L', 'V', 'Z', 'R', 'H', 'Q'],
    '9': ['B', 'H', 'G', 'N', 'F', 'Z', 'L', 'D'] 
} 

for direction in day_five_list:
    direction_list = direction.split()
    if direction_list[0] == 'move':
        quantity = int(direction_list[1])
    if direction_list[2] == 'from':
        start_key = direction_list[3]
    if direction_list[4] == 'to':
        end_key = direction_list[5]

    moved_crates = crates_dict[start_key][-1*quantity:]

    ##Part 2 comment this line
    moved_crates.reverse()
    
    crates_dict[start_key] = crates_dict[start_key][:-1*quantity]
    crates_dict[end_key] = crates_dict[end_key] + moved_crates


    print(direction)
    print(moved_crates)
    print(crates_dict)
    
    
```
