```python
day_four_input = open('day_4_input.txt').read()
day_four_list = day_four_input.splitlines()
```

```python
counter = 0
for pair in day_four_list:
    first_pair = pair.split(',')[0]
    second_pair = pair.split(',')[1]
    first_start = int(first_pair.split('-')[0])
    first_end = int(first_pair.split('-')[1])
    second_start = int(second_pair.split('-')[0])
    second_end = int(second_pair.split('-')[1])

    if (first_start <= second_start and first_end >= second_end) or (second_start <= first_start and second_end >= first_end):
        print(pair, 'T')
        counter += 1
    else:
        print(pair, 'F')
  
```
